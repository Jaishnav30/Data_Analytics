```python

import matplotlib.pyplot as plt
import numpy as np

# Given data
N = np.array([5000, 50000, 500000])

# Measured execution times (in seconds)
best_case = np.array([0.0367, 2.7723, 276.1640])
worst_case = np.array([0.0356, 2.6741, 269.8167])
avg_case = np.array([0.0628, 2.8255, 282.8167])

# O(n^2) reference curve (scaled for comparison)
big_o_n2 = (N**2) / (10**9)  # Adjust scale for visibility

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(N, best_case, 'go-', label="Best Case")
plt.plot(N, worst_case, 'ro-', label="Worst Case")
plt.plot(N, avg_case, 'bo-', label="Average Case")
plt.plot(N, big_o_n2, 'k--', label="O(n²)")

# Labels and title
plt.xlabel("Input Size (N)")
plt.ylabel("Time Taken (seconds)")
plt.title("Selection Sort Time Complexity")
plt.legend()
plt.grid(True)

# Show the graph
plt.show()
```
